# Вопрос 1 - правильно
При осуществлении образовательной деятельности по образовательным программам высшего образования организация обеспечивает:
- $+$ реализацию дисциплин (модулей) (включая проведение текущего контроля успеваемости); проведение практик; проведение промежуточной аттестации обучающихся; проведение итоговой (государственной итоговой) аттестации обучающихся
- $-$ реализацию дисциплин (модулей); проведение практик; проведение итоговой (государственной итоговой) аттестации обучающихся
- $-$ реализацию дисциплин (модулей); проведение практик; проведение промежуточной аттестации обучающихся; проведение итоговой (государственной итоговой) аттестации обучающихся
- $-$ реализацию дисциплин (модулей) (включая проведение текущего контроля успеваемости); проведение промежуточной аттестации обучающихся

# Вопрос 2 - правильно
Научный или научно-технический проект это:
- $-$ деятельность, направленная на получение, применение новых знаний для решения технологических, инженерных, экономических, социальных, гуманитарных и иных проблем, обеспечения функционирования науки, техники и производства как единой системы
- $-$ научный и (или) научно-технический результат, содержащий новые знания или решения и зафиксированный на любом информационном носителе
- $+$ комплекс скоординированных и управляемых мероприятий, которые направлены на получение научных и (или) научно-технических результатов и осуществление которых ограничено временем и привлекаемыми ресурсами

# Вопрос 3 - правильно 
Программы бакалавриата и программы специалитета реализуются образовательными организациями
- $+$ высшего образования
- $-$ высшего образования и научными огранизациями
