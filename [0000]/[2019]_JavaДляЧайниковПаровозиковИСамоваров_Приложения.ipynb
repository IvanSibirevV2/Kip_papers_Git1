# Загрузка и установка JDK
Если вы работаете в операционной системе Windows, Linux или Solaris, загрузите пакет JDK (Java Development Kit -- комплект разработчика приложений на языке Java), посетив официальный сайт компании Oracle (http:// oracle.com ). Пользователям Mаc предоставляются другие возможности.

# Java на платформе Windows, Linux и Solaris
Для загрузки JDK перейдите на сайт компании Oracle по следующему адресу: http://oracle.com/technetwork/java/javase/downloads

Щелкнув на кнопке DOWNLOAD, вы перейдете на страницу со списками вариантов загрузки JDK для различных платформ. В любом случае учитывайте следующее.

- Компания Oracle выпускает три варианта платформы Java: SE, ЕE и МЕ. На всех этих платформах используется один и тот же язык Java, но каждая из них имеет свою стандартную библиотеку или API (Application Programming Interface - интерфейс прикладного программирования). Загрузив Java, вы получаете определенный набор инструментов для работы с исходным кодом и одну из трех библиотек. Ниже приведено краткое описание каждого варианта платформы Java.

- Java SE (Standard Edition - стандартная версия) содержит все, что необходимо для работы на настольном компьютере, включая обработкутекста, математические вычисления, ввод и вывод данных, коллекции объектов и т.п. Для работы с примерами данной книги достаточно установить платформу Java SE.
- Java EE (Enterprise Edition - корпоративная версия) содержит средства, необходимые для работы с серверами средних и крупных предприятий. В нее включены инструменты для работы с веб-серверами и базами данных, средства обмена сообщениями между клиентами и серверами, средства управления распределенными системами и т.п.
- Java ME (Micro Edition - компактная версия) предназначена для мобильных устройств, таких как смартфоны, планшеты, плееры и пр. Эта версия платформы предоставляет ограниченный набор средств, необходимых для специализированных устройств, менее мощных, чем настольные компьютеры. Для работы с данной книгой вам необходимо издание SE.
- Существуют разные версии пакета Java, и каждая из них имеет ряд обновлений. Когда-то номера версий Java начинались с единицы. Первой была версия 1.0, затем появилась версия 1.1, после чего общепринятый порядок нумерации версий был нарушен: появилась Java 2, v.1.2, после нее - Java 2, v.1.3 и т.д. Почему произошел скачок от 1.4 до 5.0, до сих пор никто не может объяснить. В конце концов, первое число решили убрать, и появились версии Java 6, Java 7, Java 8 и Java 9 (на данный момент это последняя версия). Сразу же после выпуска новой версии поставляется ее обновление номер 1 (update 1, или u1). Постепенно номер обновления увеличивается. На данный момент поставляется версия Java SE 9.0.1.

- Существуют два типа загрузок Java. На сайте Oracle вы увидите в каждой версии и в каждом издании два пакета: JRE (Java Runtime Environment - среда выполнения Java) и JDK (Java Development Kit - комплект разработчика приложений на языке Java). В таком случае отдельная загрузка пакета JRE вам не нужна.

Вы можете бесплатно загрузить интегрированную рабочую среду NetBeans, средства JavaFX, исходные коды Java, документацию и другие материалы. Остальные материалы нужны только для решения специальных задач, например для создания сценариев JavaFX на веб-страницах.

В Linux инсталляционный файл имеет расширение .rpm или .bin. Установите его на компьютере с помощью менеджера пакетов Linux.

# Java для Macintosh
В первую очередь, нужно выяснить, установлен ли пакет JDK на компьютере. Это можно сделать одним из двух способов.

# Установка Eclipse
Процесс установки Eclipse зависит от операционной системы и типа файла, который вы загрузили.
- Вы работаете в Windows и загрузили саморазворачивающийся архив .еxe. Дважды щелкните в Проводнике Windows на файле . exe и разверните файл в любой папке. Рекомендуется скопировать содержимое в папку Program Files, в которой находятся остальные программы.
- Вы работаете в Windows и загрузили файл архива . zip. Извлеките содержимое архива в любую папку. Для этого откройте Проводник Windows, найдите загруженный файл . zip и дважды щелкните на нем. Архив откроется в программе архивации. Перетащите папку eclipse в любое место файловой системы. Как и в предыдущем случае, рекомендуется перетащить ее в папку Program Files. Инсталлятор и системный реестр Windows при установке Eclipse не используются. Развернув содержимое архива в любой папке на жестком диске, можете сразу запустить программу Eclipse. Для этого дважды щелкните на файле eclipse. exe.
- Вы работаете в Мaс OS X. Если вы загрузили файл . tar.gz, найдите его в папке Downloads и дважды щелкните на нем. Двойной щелчок приведет к извлечению содержимого архива. После этого в папке Downloads должна появиться папка Eclipse. Перетащите ее в папку Applications. Если вы загрузили файл . dmg, ваш браузер должен открыть его. В противном случае найдите файл . dmg в папке Downloads и дважды щелкните на нем. Выполняйте инструкции, выводимые на экране. После этого перетащите папку Eclipse в папку Applications.
- Вы работаете в Linux. Вы загрузили либо архив . tar. gz, либо саморазворачивающийся архив . bin. Извлеките содержимое архива или дважды щелкните на саморазворачивающемся архиве.

4. Щелкните на значке Workbench, чтобы открыть главный экран Eclipse. Будет активизирован главный экран Eclipse с заданным рабочим пространством. На панели Package Explorer (Проводник пакетов) откройте папку с примерами данной книги (рис. А.1).

Замена инструкции username != null && password != null
инструкцией ! (username == null | | password == null) Программа остается работоспособной, поскольку инструкция username ! = null && password != null логически эквивалентна инструкции ! (username == null || password == null). Эта эквивалентность вытекает из логического закона де Моргана. В данном случае применяется дистрибутивный закон де Моргана для операторов !, && и | |.

Замена инструкцииusername != null && password != null
инструкцией ! (username == null & password == null) Программа аварийно завершается, как только пользователь щелкает на кнопке Cancel для одного из вызовов JOptionPane.showInputDialog. Причина заключается в том, что условие ! (username == null && password == null) не является логически эквивалентным условию исходной программы. Предположим, например, что пользователь ввел bburd в диалоговом окне Имя пользователя, но щелкнул на кнопке Cancel в диалоговом окне Пароль. В результате значение username будет равным bburd, а значение password --- null (в данном случае password не содержит корректного значения String). Тогда условие ! (username == null && password == null) будет равно true, поскольку это не тот случай, когда обе переменные, username и password, равны null. Итак, Java продолжает проверять значение выражения username. equals("bburd"). Затем Java переходит к проверке значения выражения password.equals("swordfish"). Но попытка вызова метода equals для переменной password приводит к аварийному завершению программы, поскольку переменная password не содержит корректного значения String. # Замена FlowLayout на BorderLayout
Кнопка растягивается на весь фрейм. В Java используется несколько диспетчеров макетов, которые ведут себя по-разному. В классе FlowLayout размеры компонентов уменьшаются в соответствии с размерами содержащегося в них контента. В классе BorderLayout компоненты расширяются, заполняя включающий их компонент. Существуют и другие различия между диспетчерами макетов. Например, в классе FlowLayout объекты выравниваются по горизонтали. В классе BorderLayout объекты отображаются либо в центре, либо в одном из четырех углов макета.

# Описание использования ключевого слова this в примере кода
Первое использование ключевого слова this приводит к тому, что выражение this.n ссылается на поле n:

    private int n;
а не на параметр n:

    IntegerHolder(int n)
Второй случай использования ключевого слова this относится к экземпляру класса IntegerHolder (в зависимости от того, какой экземпляр класса выполняет свой метод displayMyN в данный момент времени). Этот экземпляр передается методу display класса Displayer. В тело метода display этот экземпляр передается параметром holder. Метод display вызывает метод getN объекта holder. Другими словами, метод display вызывает метод getN экземпляра класса IntegerHolder.
